```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
```

```python
image_path = 'D:/TGMT/04.jpg'
image = Image.open(image_path)
```

```python
# nh xm
gray_image = image.convert('L')

# Chuyn nh xm v mng
image_a = np.array(gray_image)

# Lm m nh xm vi b lc Gaussian
blurred_image_a = ndimage.gaussian_filter(image_a, sigma=6, mode="constant")
blurred_image_b = ndimage.gaussian_filter(image_a, sigma=9, mode="constant")
blurred_image_c = ndimage.gaussian_filter(image_a, sigma=12, mode="constant")

# Hin th nh
fig, axs = plt.subplots(2,2, figsize=(10, 8))

plt.subplot(2,2,1)
plt.imshow(gray_image, cmap='gray')
plt.title('nh xm gc')

plt.subplot(2,2,2)
plt.imshow(blurred_image_a, cmap='gray')
plt.title('sigma=6')

plt.subplot(2,2,3)
plt.imshow(blurred_image_b, cmap='gray')
plt.title('sigma=9')

plt.subplot(2,2,4)
plt.imshow(blurred_image_c, cmap='gray')
plt.title('sigma=12')
```
